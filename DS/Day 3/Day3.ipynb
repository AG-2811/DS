```python
def student_records(**kwargs):
    import pandas as pd
    import numpy as np
    try:
        data = pd.DataFrame(kwargs)
        data.index = np.arange(1, len(data) + 1)
        return data
    except:
        data = pd.DataFrame(kwargs,index=[1])
        return data
```

```python
student_records(Name=['Ajay','Anuj','Aman'],
                Age=['',18,19],
                Address=['Noida','Delhi',''],
               Sec = list('A'' ''C'))
```

```python
def show_time():
    try:
        while True:
            import time
            print(time.asctime())
            time.sleep(1)
            display(clear = True)
    except KeyboardInterrupt:
        print('Timer has been stopped!!')
        
        
```
