```python
from googlesearch import search
data = search('IITM',num_results =10)
all_data=[]
for i in data:
    all_data.append(i+'\n')
file = open('Data_Science.txt',mode='w')
file.writelines(all_data)
file.close()
print('Done')
```

```python
from googlesearch import search
title=input('Enter Your Search: ')
data = search(title,num_results =10)
all_data=[]
for i in data:
    all_data.append(i+'\n')
# print(all_data)
try:
    file = open(f'{title}.txt',mode='x')
    #mode x: it only create new file
    file.writelines(all_data)
    file.close()
    print('Done')
except: 
    print('File is already exist')
```

```python
url = r'''https://preview.redd.it/shinchan-wallpaper-v0-bi1ma4oqprtd1.jpeg?width=640&crop=smart&auto=webp&s=50f5bde5c36fef06899ef3a9e1748fead668a8b6'''
import requests as r
image_data = r.get(url).content
f = open('download_image.jpg','wb')
f.write(image_data)
f.close()
from IPython import display
display.Image('download_image.jpg')
```

```python
url ='''https://picsum.photos/1000/1080'''
import requests as r
no = 10
for i in range(no):
    img_data = r.get(url).content
    f = open(f'image_{i}.jpeg','wb')
    f.write(img_data)
    f.close()
    from IPython import display as d
    display(d.Image(f'image_{i}.jpeg',width=200,height=200))
```
